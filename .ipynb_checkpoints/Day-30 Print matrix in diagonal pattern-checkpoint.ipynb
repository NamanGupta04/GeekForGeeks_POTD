```python
class Solution:
    def matrixDiagonally(self,mat):
        # code here
        res = []
        m = 0
        n = 0
        s = len(mat)-1
        diagUp = False
        diagDown = False
        for i in range(len(mat)* len(mat)):
            res.append(mat[m][n])
            
            if m == s and n == s:
                break
            if m == 0 and n == 0:
                n = 1
                diagDown = True
                continue
            # time to update it!
            if diagUp and m == 0:
                if n < s:
                    n += 1
                    diagUp = False
                    diagDown = True
                    
                else:
                    m += 1
                    diagDown = True
                    diagUp = False
                continue
            
            if diagDown and n == 0:
                if m < s:
                    m+=1
                    diagUp = True
                    diagDown = False
                
                else:
                    n += 1
                    diagUp = True
                    diagDown = False
                continue
            
            
            if diagDown and m == s:
                n += 1
                diagUp = True
                diagDown = False
                continue
            if diagUp and n == s:
                m += 1
                diagDown = True
                diagUp = False
                continue
            
            if diagUp:
                m -= 1
                n += 1
                continue
            
            if diagDown:
                m+=1
                n -= 1
                continue
                    
        return res
```
